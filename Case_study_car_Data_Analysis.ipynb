# Cars Dataset Visualizations

# bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(8, 8))
ax = sns.countplot(x="Buyer Gender",data=df)
plt.title("Male vs Female buyers participated in the sales");
for p in ax.patches:
    ax.annotate(p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
```

```python
genderCnt=df["Buyer Gender"].value_counts()
chrt=genderCnt.plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Male vs Female buyers participated in the sales");
```

# Find the top 5 cars based on their sales price.

```python
df.sort_values(by=["Sale Price"])
```

```python
df.sort_values(by=["Sale Price"]).tail(5)[["Make","Model","Nickname"]]
```

```python
carBySalePrice=df.sort_values(by=["Sale Price"]).tail(5)
plt.style.use("seaborn")
plt.plot(carBySalePrice["Make"]+'-'+carBySalePrice["Model"],carBySalePrice["Sale Price"])
plt.xlabel("Car Model",size="14")
plt.ylabel("Sale Price",size="14")
plt.title("Top 5 cars based on sale price",size="16");
plt.tight_layout()
```

```python
#Finding Top 5 cars based on average sale price
```

```python
x=df.groupby(["Make","Model"])["Sale Price"].mean()
type(x)
x=x.to_frame()
x=x.rename(columns={"Sale Price": "Avg Sale Price"})
x.sort_values(by=["Avg Sale Price"]).tail(5)
```

# Find the least 5 cars based on their Resell price.

```python
df.sort_values(by=["Resell Price"])
```

```python
df.sort_values(by=["Resell Price"]).head(5)[["Make","Model","Nickname"]]
```

```python
reSellCar=df.sort_values(by=["Resell Price"]).head(5)
plt.style.use("seaborn")
plt.plot(reSellCar["Make"]+'-'+reSellCar["Model"],reSellCar["Resell Price"])
plt.xlabel("Car Model",size="14")
plt.ylabel("Resell Price",size="14")
plt.title("Least 5 cars based on resell price",size="16");
plt.tight_layout()
```

```python
#Finding least 5 cars based on average resell price
```

```python
x=df.groupby(["Make","Model"])["Resell Price"].mean()
x=x.to_frame()
x=x.rename(columns={"Resell Price": "Avg Resell Price"})
x.sort_values(by=["Avg Resell Price"]).head(5)
```
