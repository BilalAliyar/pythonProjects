# Iris Dataset Visualizations

```python
#Creating scatter plot for the possible pairs of PL,PW,SL and SW columns for each classification
```

```python
setosaData=df[df["Classification"]=="Iris-setosa"]
versicolorData=df[df["Classification"]=="Iris-versicolor"]
virginicaData=df[df["Classification"]=="Iris-virginica"]

plt.figure(figsize=(13, 10))
plt.suptitle("Iris data distribution",size=20,color="black")
plt.style.use("seaborn")
plt.subplot(221)
plt.scatter(setosaData['PW'],setosaData['PL'],label="Iris-setosa")
plt.scatter(versicolorData['PW'],versicolorData['PL'],color="black",label="Iris-versicolor")
plt.scatter(virginicaData['PW'],virginicaData['PL'],color="green",label="Iris-virginica")
plt.title(" Petal Width vs Petal Length", size=14,color="brown")
plt.ylabel("Petal Length")
plt.xlabel("Petal Width")
#plt.legend(fontsize="10")

plt.subplot(222)
plt.style.use("seaborn")
plt.scatter(setosaData['SW'],setosaData['SL'],label="Iris-setosa")
plt.scatter(versicolorData['SW'],versicolorData['SL'],color="black",label="Iris-versicolor")
plt.scatter(virginicaData['SW'],virginicaData['SL'],color="green",label="Iris-virginica")#plt.scatter(df['Classification'],df['SL'], color="red")
plt.title(" Sepal Width vs Sepal Length", size=14,color="brown")
plt.ylabel("Sepal Length")
plt.xlabel("Sepal Width")
#plt.legend(fontsize="10")

plt.subplot(223)
plt.scatter(setosaData['SW'],setosaData['PW'],label="Iris-setosa")
plt.scatter(versicolorData['SW'],versicolorData['PW'],color="black",label="Iris-versicolor")
plt.scatter(virginicaData['SW'],virginicaData['PW'],color="green",label="Iris-virginica")#plt.scatter(df['Classification'],df['SL'], color="red")
plt.title(" Sepal Width vs Petal Width", size=14,color="brown")
plt.ylabel("Petal Width")
plt.xlabel("Sepal Width")
#plt.legend(fontsize="10")

plt.subplot(224)
plt.style.use("seaborn")
plt.scatter(setosaData['SL'],setosaData['PL'],label="Iris-setosa")
plt.scatter(versicolorData['SL'],versicolorData['PL'],color="black",label="Iris-versicolor")
plt.scatter(virginicaData['SL'],virginicaData['PL'],color="green",label="Iris-virginica")#plt.scatter(df['Classification'],df['SL'], color="red")
plt.title(" Sepal Length vs Petal Length", size=14,color="brown")
plt.ylabel("Petal Length")
plt.xlabel("Sepal Length")
plt.legend(loc="lower right" ,bbox_to_anchor=(1.35, 0),fontsize="12")

#plt.subplots_adjust(right=0.25)

plt.tight_layout()

```

```python
#Creating boxplot for the distribution of PL,PW,SL and SW values against each classification
```

```python
plt.figure(figsize=(12, 10))
plt.subplot(221)
sns.boxplot(x="Classification", y="PL",data=df)
plt.ylabel("Petal Length")
plt.subplot(222)
sns.boxplot(x="Classification", y="PW",data=df)
plt.ylabel("Petal Width")
plt.subplot(223)
sns.boxplot(x="Classification", y="SL",data=df)
plt.ylabel("Sepal Length")
plt.subplot(224)
sns.boxplot(x="Classification", y="SW",data=df)
plt.ylabel("Sepal Width")

plt.tight_layout()
```

```python
#Creating histograms for the distribution of PL,PW,SL and SW values
```

```python
plt.figure(figsize=(12, 10))
plt.subplot(221)
plt.title("Distribution of Petal Length")
df["PL"].hist()
plt.subplot(222)
plt.title("Distribution of Petal Width")
df["PW"].hist()
plt.subplot(223)
plt.title("Distribution of Sepal Length")
df["SL"].hist()
plt.subplot(224)
plt.title("Distribution of Sepal Width")
df["SW"].hist()
plt.tight_layout()
```

```python
#creating pair plot using seaborn for the distribution of PL,PW,SL and SW values for each classification
```

```python
#Renaming the columns
df=df.rename(columns={"PL": "Petal Length", "PW": "Petal Width", "SL": "Sepal Length", "SW": "Sepal Width"})
sns.pairplot(df,hue="Classification",height=3,markers=["o", "s", "D"]);
```

```python
#Creating violin plot for the distribution of PL,PW,SL and SW values against each classification
```

```python
plt.figure(figsize=(12, 10))
plt.subplot(221)
sns.violinplot(x="Classification", y="Petal Length",data=df)
plt.subplot(222)
sns.violinplot(x="Classification", y="Petal Width",data=df)
plt.subplot(223)
sns.violinplot(x="Classification", y="Sepal Length",data=df)
plt.subplot(224)
sns.violinplot(x="Classification", y="Sepal Width",data=df)


plt.tight_layout()
```
